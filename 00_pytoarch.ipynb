### Reshaping, stakcing, squeezing and unsquezzing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but keep the same memoory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - remove all 1 dim from a tensor
* Unsqueeze - add a 1 dim to a target tesnor
* Permute - return a view of the input with dimesnion permuted(swapped) in a certain way


```python
# Let's create a tensor
x = t.arange(1. ,11.)
x, x.shape
```

```python
# Add an extra dimeions(reshape), needs to be compatible wtih the og shape
x_reshaped = x.reshape(1,10) # could be (10,1), could be (2,5)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 10)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor share the same memory as teh og input)
z[:, 0] = 5
z, x
```

```python
#Stack tensors on top of each other
x_stacked = t.stack([x, x, x, x], dim = 0)
x_stacked
```

```python
x_stacked = t.stack([x, x, x, x], dim = 1)
x_stacked
```

```python
# Verticaly stacks the tensors on top each otehr in diff []
x_stacked = t.vstack([x,x,x,x])
x_stacked, x_stacked.shape
```

```python
# adds tensors one after another in one []
x_stacked = t.hstack([x,x,x,x])
x_stacked, x_stacked.shape
```

```python
# torch.squeeze --> return a tensor with all of the dimenzion size one removed
x_reshaped, x_reshaped.shape
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed, x_squeezed.shape
```

```python
# torch.unsqueeze() - adds a single dim to a target tesnor at a specific dim
x_unsqueezed = x_squeezed.unsqueeze(dim=0) # dim = x, where the dim will be added
x_unsqueezed, x_unsqueezed.shape
```

```python
# torch.permute - rearranges the dim of a target tensor in a specified order (popular with immages)
x_original = t.rand(size=(224,244,3))

# Permute the og tensor to rearrange the axis(or dim) order
x_permuted = x_original.permute(2,0,1) # shifted axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
# since this is just different view it is stored in the same memory
```

### Reproducbility (trying to take random out of random)

In short how a nn learns:

1. Start with random numbers
2. Tensor operations
3. Update random numbers to try and make them better represnetation of the data
4. again
5. again
6. again

To reduce the randomness in nn and PyTorch comes the concept of a **random seed**.

Essentailly what the random seed does is "flavour" the randomness.

```python
# Create two random tensors
rand_tensor_a = t.rand(3,4)
rand_tensor_b = t.rand(3,4)

print(rand_tensor_a)
print(rand_tensor_b)
print(rand_tensor_a == rand_tensor_b)
```

```python
# Lets make some random byt reproducible tensors
import torch as t

# Set the random seed
RANDOM_SEED = 42 #0, 1234 are popoular
t.manual_seed(RANDOM_SEED) # only works for one radomness
rand_tensor_C = t.rand(3,4)

t.manual_seed(RANDOM_SEED)
rand_tensor_D = t.rand(3,4)

print(rand_tensor_C)
print(rand_tensor_D)
print(rand_tensor_C == rand_tensor_D)
```
